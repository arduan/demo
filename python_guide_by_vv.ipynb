**<h2>3. Плюсы и минусы Python</h2>**

Итак, ты дочитал уже до этого пункта. Молодец!
Теперь, прежде чем ты окончательно решишь связать свою жизнь с программированием и языком программирования (<b>ЯП</b>) Python в частности, пожалуйста, посмотри на его "+" и "-". Это позволит понять, подходит ли он под твои цели и нужно ли на него тратить свое время.

Преимущества Python:

1. Простой синтаксис, который похож на английский язык;
2. Динамическая типизация (это как плюс так и минус, но я лично считаю плюсом);
3. Большое мировое сообщество и огромное количество библиотек. Шансы, что у вас возникнет вопрос, который еще не решен - минимальны, особенно в начале вашего пути в программировании.
4. Отлично подходит для прототипирования. Можно за 5-10 минут сделать прототип, который в других ЯП займет несколько часов.
5. Широко используется в веб-деве, создании ботов и парсеров и в data science.

Недостатки Python:
1. Медленее, чем некоторые другие ЯП;
2. Проблемы с тредингом;
3. Не нативный для мобильной среды, что затрудняет разработку мобильных приложений;
4. Требует значительных объемов памяти.

Подводя итоги, Python хорошо подходит начинающим программистам из-за своего легкого синтаксиса. Кроме того, на нём удобно писать сравнительно небольшие программы и скрипты, которые подойдут для персонального использования или малого/среднего бизнеса. Если ты мечтаешь построить программу, которой будет пользоваться больше 100 тысяч людей каждый день, то лучше начни учить другой ЯП, который больше подходит к таким задачам.

**<h2>8. Как правильно решать задачи в программировании</h2>**

Тебе, как начинающему программисту, конечно же хочется побыстрее всё выучить и начать зарабатывать деньги. Я считаю, что для того, чтобы перейти в категорию джуна, то нужно потратить около 1000 часов на учёбу. Проблема заключается в том, что эти 1000 часов можно потратить с разной эффективностью, что скажется на твоем опыте и знаниях после окончания учебы. Тут стоит уточнить, что речь идёт об учебе, чтобы стать джуном, так как в программировании учёба не заканчивается никогда. Все время придётся учить новые фреймворки, ЯП и тд, чтобы быть конкурентноспособным. 
Так вот, в интернете есть множество статей, как правильно решать задачи в программировании. Например, ты можешь прочитать эту статью [10 Steps to Solving a Programming Problem](https://codeburst.io/10-steps-to-solving-a-programming-problem-8a32d1e96d74).
Лично я считаю, что любой начинающий программист должен стараться по-максимуму делать задания самостоятельно (использовать либы не запрещается, а вот искать сразу готовое решение - не лучший выбор для новичка). Поэтому, я советую придерживаться следующего алгоритма решения задач во время учебы:
1. Понять суть задачи. Тут надо прочитать несколько раз условие, чтобы не пропустить нюансы и быть увереным, что задача понята правильно.
2. Постараться придумать алгоритм решения задачи. Если алгоритм тяжело придумать в уме, то лучше его записывать. 
3. Если ты потратил больше часа времени на алгоритм, но так и не придумал решение, то можно загуглить именно алгоритм решения такой задачи (не надо гуглить готовое решение).
4. Написать код, который реализует придуманный/найденный алгоритм. Если написать код тяжело, то сначала напиши псевдокод, а потом перепиши его в нормальный код (тут можно некоторые части гуглить)
5. Если задача слишком сложная и не получается написать ее решение, то можно загуглить готовое решение, но обязательно нужно его разобрать. Также, лучше остановись на задачах этого уровня сложности и решай их, пока не поймешь, что они легко даются тебе.
6. Поздравляю! Ты написал или нашел готовый код решения задачи. Посмотри на него и подумай, что можно улучшить. Проверь, что он действительно правильно работает и возвращает правильные ответы для разных типов входных данных. 
7. Почитай подробнее про вещи, с которыми у тебя возникли проблемы в коде.

PS. Очень часто начинающие программисты сразу бегут с вопросами к более опытным программистам вместо того, чтобы сесть и подумать самим. Я считаю, что это абсолютно неверная тактика, так как умение думать и находить решения  самостоятельно (может быть даже с помощью Гугла) является **САМЫМ НЕОБХОДИМЫМ НАВЫКОМ** для любого программиста. Поэтому, старайся просить помощи у других людей только после того, как ты сам подумал, погуглил, попытался написать код, подумал, что не так с кодом, и попробовал исправить. Если ты всё это сделал и всё равно не получается, тогда проси помощи.
